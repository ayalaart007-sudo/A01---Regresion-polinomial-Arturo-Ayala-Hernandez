```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df1 = pd.read_csv('bateria_tableta.txt')
x_bat = np.array(df1['carga'])
y_bat = np.array(df1['duracion'])
```

```python
#g1
X1 = np.array([[v] for v in x_bat])
Lr1 = LinearRegression()
Lr1.fit(X1, y_bat)

#g3
X3 = np.array([[v, v**2, v**3] for v in x_bat])
Lr3 = LinearRegression()
Lr3.fit(X3, y_bat)

#g5
X5 = np.array([[v, v**2, v**3, v**4, v**5] for v in x_bat])
Lr5 = LinearRegression()
Lr5.fit(X5, y_bat)

#g7
X7 = np.array([[v, v**2, v**3, v**4, v**5, v**6, v**7] for v in x_bat])
Lr7 = LinearRegression()
Lr7.fit(X7, y_bat)
```

```python
print("--- COMPARACIÓN R2 BATERÍA ---")
print("Grado 1 - R2:", Lr1.score(X1, y_bat))
print("Grado 3 - R2:", Lr3.score(X3, y_bat))
print("Grado 5 - R2:", Lr5.score(X5, y_bat))
print("Grado 7 - R2:", Lr7.score(X7, y_bat))

print("\nAtributos Grado 7:")
print("Intercepto:", Lr7.intercept_)
print("Coeficientes:", Lr7.coef_)
```

```python
plt.scatter(x_bat, y_bat, color='black', label='Datos')

x_ordenado = np.sort(x_bat)

plt.plot(x_ordenado, Lr1.predict(np.array([[v] for v in x_ordenado])), label='Grado 1')
plt.plot(x_ordenado, Lr3.predict(np.array([[v, v**2, v**3] for v in x_ordenado])), label='Grado 3')
plt.plot(x_ordenado, Lr5.predict(np.array([[v, v**2, v**3, v**4, v**5] for v in x_ordenado])), label='Grado 5')
plt.plot(x_ordenado, Lr7.predict(np.array([[v, v**2, v**3, v**4, v**5, v**6, v**7] for v in x_ordenado])), label='Grado 7')

plt.legend()
plt.title("Modelos Batería")
plt.show()
```

```python
df2 = pd.read_csv('Salarios_minimos.csv')
x_sal = np.array(df2['Periodo'])
y_sal = np.array(df2['Salario'])

#g1
XS1 = np.array([[v] for v in x_sal])
LrS1 = LinearRegression().fit(XS1, y_sal)

#g3
XS3 = np.array([[v, v**2, v**3] for v in x_sal])
LrS3 = LinearRegression().fit(XS3, y_sal)

#g5
XS5 = np.array([[v, v**2, v**3, v**4, v**5] for v in x_sal])
LrS5 = LinearRegression().fit(XS5, y_sal)

#g7
XS7 = np.array([[v, v**2, v**3, v**4, v**5, v**6, v**7] for v in x_sal])
LrS7 = LinearRegression().fit(XS7, y_sal)

#Comparizon R2
print("\n--- COMPARACION R2 SALARIOS ---")
print("Grado 1:", LrS1.score(XS1, y_sal))
print("Grado 3:", LrS3.score(XS3, y_sal))
print("Grado 5:", LrS5.score(XS5, y_sal))
print("Grado 7:", LrS7.score(XS7, y_sal))

#grafica
plt.figure()
plt.scatter(x_sal, y_sal, color='black', label='Datos')
plt.plot(x_sal, LrS1.predict(XS1), label='Grado 1')
plt.plot(x_sal, LrS3.predict(XS3), label='Grado 3')
plt.plot(x_sal, LrS5.predict(XS5), label='Grado 5')
plt.plot(x_sal, LrS7.predict(XS7), label='Grado 7')
plt.legend()
plt.title("Modelos Salarios")
plt.show()
```
